```python
import pandas as pd
```

```python
df = pd.read_csv('./base.csv', header=None)
df = df.rename(columns={
    0 : 'kode_lokasi',
    1 : 'nama_lokasi'
})
df
```

```python
df_dropped = df[df['kode_lokasi'].apply(lambda x : len(str(x)) >= 13)]
df_dropped['nama_lokasi'] = df_dropped['nama_lokasi'].str.lower()
df_dropped
```

```python
df_dropped_1 = df_dropped[df_dropped['nama_lokasi'].duplicated()]
df_dropped_1.index
```

```python
test_df = df_dropped.drop(index=df_dropped_1.index)
test_df
```

```python
test_df.set_index('nama_lokasi').to_json('./loc_code.json')
```
